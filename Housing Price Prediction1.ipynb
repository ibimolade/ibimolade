```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
#Load housing1 dataset
df= pd.read_csv('housing1.csv')
```

```python
#Print last 5 rows of housing dataset
df.tail()
```

```python
#Check to see which column has missing data
df.isnull().sum()
```

```python
#fill in the empty data cells with 0
df.fillna(0, inplace=True)
```

```python
#Check to see all data cells have been filled
df.isnull().sum()
```

```python
#Conver categorical data to numerical data
df= pd.get_dummies(df, columns=[])
```

```python
df.tail()
```

```python
#Extract column that will be used to predict prices by
a= df.drop("median_house_value", axis=1)
b= df["median_house_value"]
```

```python
#Test(80%) and train(20%) the computer by splitting the data
a_train,a_test,b_train,b_test= train_test_split(a,b, test_size=0.2, random_state=21)
```

```python
#Create the Linear Regression model used to predict the prices
model= LinearRegression()
model.fit(a_train, b_train)
```

```python
#Prediction
b_pred= model.predict(a_test)
```

```python
#Evaluate the error margin of the computer
mae=mean_absolute_error(b_test, b_pred)
mse=mean_squared_error(b_test, b_pred)
rmse= mse**0.5
```

```python
#Print the error margin
print(f"mae: {mae}, rmse:{rmse}")
```
